#1. Membuat DataFrame

```python
import pandas as pd
import numpy as np

# Data penjualan dari gambar yang telah diidentifikasi
data = [100, 150, 50, 100, 130, 120, 100, 85, 70, 150,
                   150, 120, 50, 100, 100, 140, 90, 150, 50, 90,
                   120, 100, 110, 75, 65]

# Membuat DataFrame 1 kolom dengan 25 baris
df = pd.DataFrame(data, columns=['TransactionAmount'])

df
```

#2. Analisis Measure of Central Tendency

```python
# Analisis measure of central tendency
mean_value = df['TransactionAmount'].mean()
median_value = df['TransactionAmount'].median()
mode_value = df['TransactionAmount'].mode()[0]

mean_value, median_value, mode_value

```

#3. Analisis Measure of Variability

```python
# Analisis measure of variability
variance_value = df['TransactionAmount'].var()
std_dev_value = df['TransactionAmount'].std()
range_value = df['TransactionAmount'].max() - ['TransactionAmount'].min()
quartiles_value = df['TransactionAmount'].quantile([0.25, 0.5, 0.75])


variance_value, std_dev_value, range_value, quartiles_value
```

#5. Menentukan H0 dan H1 T-Test

H0 : Penjualan rata-rata perusahaan X = $100

H1 : Penjualan rata-rata perusahaan X > $100


#Melakukan T-Test:

```python
from scipy import stats

# Melakukan T-Test
t_stat, p_value = stats.ttest_1samp(df['TransactionAmount'], 100)

t_stat, p_value
```

Dengan hasil T-Test sebagai berikut:

*  Nilai t-statistik (t_stat): 0.408
*  Nilai p-value (p_value): 0.687

**Interpretasi Hasil:**

P-value:

*   Nilai p-value sebesar 0.687 jauh lebih besar dari tingkat signifikansi yang umum digunakan (Î± = 0.05).
*   Karena p-value > 0.05, kita gagal menolak H0.

Kesimpulan:

Tidak ada bukti yang cukup untuk mengatakan bahwa rata-rata jumlah transaksi setelah training berbeda secara signifikan dari $100.
Ini berarti training yang dilakukan kepada pekerja sales mungkin tidak memiliki dampak signifikan terhadap peningkatan rata-rata penjualan.
